```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('hops.csv')
display(data.columns)
stringCols = ['Name', 'Purpose', 'Country', 'International Code', 'Ownership', 'Flavor Tags', 'Cultivar/Brand ID', 'Aroma Profile']
# filter out the string columns using slice
data[data.columns[~data.columns.isin(stringCols)]] = data[data.columns[~data.columns.isin(stringCols)]].fillna(0)
data = data.replace('Unknown', np.nan)
numericData = data.drop(stringCols, axis=1)
# create a new column called PurposeNum that converts the Purpose column to a numeric value, based on np.unique
data['PurposeNum'] = data['Purpose'].map({purpose: i for i, purpose in enumerate(np.unique(data['Purpose']))})
# Aroma Profile is a dictionary where flavor tags are the keys and the values are the intensity of the flavor
aromaProfile = data['Aroma Profile']
aromaProfile = [eval(profile) if type(profile) == str else {} for profile in aromaProfile]
# create a new dataframe from the list of dictionaries
aromaProfile = pd.DataFrame(aromaProfile)
# change the column names to be "<> Aroma"
aromaProfile.columns = [f'{col} Aroma' for col in aromaProfile.columns]
# join the dataframes
data = data.join(aromaProfile)
data['Flavor Tags'] = [eval(tags) if type(tags) == str else [] for tags in data['Flavor Tags']]
tags = data['Flavor Tags'].tolist()
# flatten the list of lists and get the unique strings
tags = np.unique([tag for sublist in tags for tag in sublist])
print(tags)
# create a new column for each called 'Flavor Tag <>' and set it to 1 if the tag is in the list of tags for that row
# Create a dictionary to hold the new columns
new_columns = {}

for tag in tags:
    new_columns[f'Flavor Tag {tag}'] = [1 if tag in tagList else 0 for tagList in data['Flavor Tags']]

# Convert the dictionary to a DataFrame
new_data = pd.DataFrame(new_columns)
display(new_data)

# Concatenate the original DataFrame with the new DataFrame
data = pd.concat([data, new_data], axis=1)
display(data)
```

```python
# print number of nans in each column
# nanCols = data.loc[:, ~data.columns.isin(stringCols)].isna().sum().sort_values(ascending=False)
# print(nanCols.head(20))
# # drop columns with more than 0 nans
# data[nanCols[nanCols == 0].index].columns.tolist()

# drop rows with more than 0 nans
num_data = data.loc[:, ~data.columns.isin(stringCols)].dropna()
data = data.iloc[num_data.index].reset_index(drop=True)
num_data = num_data.reset_index(drop=True)
display(num_data)
display(data)
```

```python
# run iterative pca
from sklearn.decomposition import IncrementalPCA

n_components = 2
ipca = IncrementalPCA(n_components=n_components, batch_size=10)
ipca.fit(num_data)
ipca_data = ipca.transform(num_data)
ipca_data = pd.DataFrame(ipca_data, columns=[f'PC{i+1}' for i in range(n_components)])
new_data = pd.concat([data[stringCols], num_data, ipca_data], axis=1)
display(new_data)
new_data.to_csv('hops_pca.csv', index=False)
```
